```python
import numpy as np
import scipy
```

```python
B_inf = np.array([[6, 2], [3, 7]])
A_inf = np.array([[1, 7], [2, 6]])

# A_inf = np.array([[1, 0], [2, 0]])  # Rank-deficient matrix
A_real = np.array([[2, 1], [1, 3]])
# B_inf = np.array([[6, 0], [3, 0]])  # Rank-deficient matrix
# B_inf = np.array([[0, 6], [0, 3]])  # Rank-deficient matrix
# A_inf = np.zeros([2, 2])
B_real = np.array([[3, 2], [2, 4]])

null_A_inf = scipy.linalg.null_space(A_inf)
span_A_inf = scipy.linalg.null_space(null_A_inf.T)
span_A_inf
```

```python
X_real_s = np.linalg.solve(
    span_A_inf.T @ np.linalg.solve(A_real, A_inf @ span_A_inf),
    span_A_inf.T @ np.linalg.solve(A_real, B_inf),
)
X_real_s
```

```python
X_inf = np.linalg.solve(A_real, B_inf - A_inf @ span_A_inf @ X_real_s)
assert ((null_A_inf @ null_A_inf.T @ X_inf) == X_inf).all()
X_inf
```

```python
null_B_inf = scipy.linalg.null_space(B_inf)
span_B_inf = scipy.linalg.null_space(null_B_inf.T)

ls_lhs = A_real @ null_A_inf
ls_rhs = (B_real - A_real @ span_A_inf @ X_real_s) @ span_B_inf @ span_B_inf.T
X_real_n, *_ = np.linalg.lstsq(ls_lhs, ls_rhs)
X_real_n
```

```python
np.array([ls_lhs @ X_real_n, ls_rhs])
```

```python
X_real = span_A_inf @ X_real_s + null_A_inf @ X_real_n
# X_real = span_A_inf @ X_real_s
X_real
```

```python
for inf_sim in [1e5, 1e10]:
    a_sim = inf_sim * A_inf + A_real
    b_sim = inf_sim * B_inf + B_real
    print(inf_sim)
    print(inf_sim * X_inf + X_real)
    print(np.linalg.solve(a_sim, b_sim))
```
